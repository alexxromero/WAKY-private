```python
# Hypothesis is a flat dist y=3
hypothesis = np.ones_like(narrow_signal)*3
```

### Narrow

```python
# -- Call the nsets method --
# extr_sets is an instance of the nsets method. If outputdir is not none, a directory will be created 
# with the data from the attributes of this class in csv files.
narrow_nsets = kwakpriv.nsets_toy(narrow_signal, hypothesis, nsets=3*10**2,
                        seed=123,
                        extrapolate=False,
                        fastGaussian=False,
                        outputdir=None)
```

```python
# Save the attributes of extr_nsets
Level = narrow_nsets.Level
Coeffs = narrow_nsets.WaveDec_data 
```

```python
r_percent=10

kwakpriv.wScalogram(data=narrow_signal, 
                hypothesis=hypothesis,
                nsigma=Coeffs,    
                nsigma_percent=r_percent*0.01,
                reconstruction_scaled=True,
                firsttrend=True,
                logscale=False,
                filled=True, 
                title="Wavelet Scalogram",
                titlesize=14,
                xlabel=None,
                textsize=12,
                textpos=0.90,
                figsize=(7,8),
                ticksize=12,
                markersize=2,
                outputfile="Wave Scal")
```
